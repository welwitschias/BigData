```python
import json
import re
from konlpy.tag import Okt
from collections import Counter
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import wordcloud
from wordcloud import WordCloud
from docutils.parsers.rst.directives import encoding
```

```python
data = json.loads(open("4차 산업혁명.json", 'r', encoding='utf-8').read())
data
```

```python
data[0]['message']
```

```python
message = ''
for i in data:
    if 'message' in i.keys():
        # 문자와 숫자가 아닌 것은 공백으로 처리
        message = message + re.sub(r'[^\w]+', ' ', i['message'])
message
```

```python
# 명사만 추출하기
nlp = Okt()
message_Noun = nlp.nouns(message)
message_Noun
```

```python
count = Counter(message_Noun)
count
```

```python
# 글자수가 2개 이상인 단어 80개 추출
word_count = dict()
for tag, counts in count.most_common(80):
    # print(tag, counts)
    if(len(str(tag))) >= 2:
        word_count[tag] = counts
word_count
```

```python
# 한글 적용
font_path = 'c:/Windows/Fonts/malgun.ttf'
font = font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rc('font', family=font)
```

```python
# 막대그래프 그리기
plt.figure(figsize=(12, 5))
sorted_Keys = sorted(word_count, key=word_count.get, reverse=True)
sorted_Values = sorted(word_count.values(), reverse=True)
plt.xlabel('키워드')
plt.ylabel('빈도수')
plt.grid(True)
plt.bar(range(len(word_count)), sorted_Values, align='center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation='90')
plt.show()
```

```python
# wordcloud 작성
wc = WordCloud(font_path='c:/Windows/Fonts/malgun.ttf', background_color='ivory',
               width=800, height=600)
cloud = wc.generate_from_frequencies(word_count)

f = plt.figure(figsize=(8, 8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
f.savefig('korword.png')
```
