```python
# 보스턴 주택가격 데이터에 머신러닝 기반의 회귀 분석

# 보스턴 주택가격 데이터
from sklearn.datasets import load_boston
import pandas as pd

boston = load_boston()
boston.DESCR
```

```python
boston.data
```

```python
boston.feature_names
```

```python
boston_df = pd.DataFrame(boston.data, columns=boston.feature_names)
boston_df
```

```python
# CRIM : 지역별 범죄 발생률
# ZN : 25,000평방피트를 초과하는 거주지역 비율
# INDUS : 비상업지역의 넓이 비율
# CHAS : 찰스강의 더미변수(1은 강의 경계, 0은 경계 아님)
# NOX : 10ppm당 일산화질소 농도
# RM : 거주할 수 있는 방의 개수
# AGE : 1940년 이전에 건축된 주택 비율
# DIS : 5개 주요 고용센터까지 가중 거리
# RAD : 고속도로 접근 용이도
# TAX : 10,000달러당 재산세 비율
# PTRATIO : 지역의 교사와 학생수 비율
# B : 지역의 흑인 거주 비율
# LSTAT : 하위 계층 비율
```

```python
# boston_df 에 PRICE 컬럼 추가
boston_df['PRICE'] = boston.target
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# X, Y {data, label(or target)} 분리하기
Y = boston_df['PRICE']
X = boston_df.drop(['PRICE'], axis=1, inplace=False)  # axis : 1은 열, 0은 행 / inplace : 대체할 것
```

```python
# 학습용(training)과 평가용(test) 데이터 분리하기
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3, random_state=156)
# test_size : 전체 데이터에서 테스트 데이터 세트의 크기를 얼마로 샘플링할 것인가?
# 0.3 = 30%(train set 70%, test set 30%) / 보통 0.3 또는 0.2 사용 / default = 0.25

# random_state : 호출할 때마다 동일한 학습용/테스트용 데이터 세트를 생성하기 위해 주어지는 난수 값
# random_state를 설정하여 수행 시 결과값을 동일하게 맞춰주는 것
# random_state에는 어떤 숫자를 적든 그 기능은 같기 때문에 어떤 숫자를 적든 상관없다
# random_state를 설정하지 않으면 수행할 때마다 다른 학습/테스트 데이터 세트가 생성(완전 무작위)
```

```python
# 선형회귀 분석모델 생성
lr = LinearRegression()

# 훈련
lr.fit(X_train, Y_train)
```

```python
# 선형회귀 분석 : 평가 데이터에 대한 "예측" 수행
Y_predict = lr.predict(X_test)
Y_predict
```

```python
# mean_squared_error(사이킷런 라이브러리) : MSE(평가지표)
# r2_scorer(사이킷런 라이브러리) : R2(결정계수, 2는 제곱수)
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
# Y_test(실제값), Y_predict(예측값)
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)  # 루트 씌워주기
```

```python
print('mse:', mse, 'rmse:', rmse)
print('R^2:', r2_score(Y_test, Y_predict))
# 결정계수(0 <= R2 <= 1) : 1에 가까울수록 회귀식 정확도가 높다
```

```python
print('Y절편값:', lr.intercept_)
print('회귀계수(기울기):', np.round(lr.coef_, 1))
```

```python
coef = pd.Series(data=np.round(lr.coef_, 2), index=X.columns)
coef.sort_values(ascending=False)
```
